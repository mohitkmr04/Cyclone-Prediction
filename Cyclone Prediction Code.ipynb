```python
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import SimpleImputer
from sklearn.metrics import accuracy_score, precision_score, recall_score, classification_report, confusion_matrix, precision_recall_curve, auc

from google.colab import drive
drive.mount('/content/drive')

# Load dataset
data = pd.read_csv("/content/drive/MyDrive/cyclone_new/atlantic.csv")

# Convert Date to datetime format
data['Date'] = pd.to_datetime(data['Date'], format='%Y%m%d')

# Convert Time to string and ensure it's in HHMM format before converting to datetime.time
data['Time'] = data['Time'].astype(str).str.zfill(4)

def convert_to_time(time_str):
    try:
        return pd.to_datetime(time_str, format='%H%M').time()
    except:
        return None

data['Time'] = data['Time'].apply(convert_to_time)

# Standardize Latitude and Longitude
def convert_latitude(lat):
    return float(lat[:-1]) * (-1 if lat[-1] == 'S' else 1)

def convert_longitude(lon):
    return float(lon[:-1]) * (-1 if lon[-1] == 'W' else 1)

data['Latitude'] = data['Latitude'].apply(convert_latitude)
data['Longitude'] = data['Longitude'].apply(convert_longitude)

# Replace -999 with NaN for better missing value handling
data.replace(-999, np.nan, inplace=True)

# Extract date-related features
data['Year'] = data['Date'].dt.year
data['Month'] = data['Date'].dt.month
data['Day'] = data['Date'].dt.day
data['DayOfWeek'] = data['Date'].dt.dayofweek
data['IsWeekend'] = data['Date'].dt.weekday >= 5

# Extract time-related features (handling possible NaN values in Time)
data['Hour'] = data['Time'].apply(lambda x: x.hour if pd.notna(x) else np.nan)
data['Minute'] = data['Time'].apply(lambda x: x.minute if pd.notna(x) else np.nan)

# Drop the original Date and Time columns after feature extraction
data.drop(columns=['Date', 'Time'], inplace=True)

# Encode categorical variables
label_encoders = {}
for column in data.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    data[column] = le.fit_transform(data[column])
    label_encoders[column] = le

# Handle missing values by imputing with median
imputer = SimpleImputer(strategy='median')
data.iloc[:, :] = imputer.fit_transform(data)

# Define features and target
X = data.drop(columns=['Status'])
y = data['Status']

# Split dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Identify top 5 important features using Random Forest
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
feature_importance = pd.Series(rf.feature_importances_, index=X.columns).sort_values(ascending=False)
top_5_features = feature_importance.index[:5]

# Define models
dt_model = DecisionTreeClassifier(min_samples_leaf=1000, criterion='entropy')
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
nb_model = GaussianNB()
ensemble_model = VotingClassifier(estimators=[('Decision Tree', dt_model), ('Random Forest', rf_model)], voting='soft')

models = {'DecisionTreeClassifier': dt_model, 'RandomForestClassifier': rf_model, 'VotingClassifier': ensemble_model, 'GaussianNB': nb_model}

def plot_confusion_matrix(y_true, y_pred, model_name, feature_set_name):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=np.unique(y_true), yticklabels=np.unique(y_true))
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.title(f'Confusion Matrix: {model_name} ({feature_set_name})')
    plt.show()

# Train and evaluate models
results = []
for feature_set, features in {'All Features': X.columns, 'Top 5 Features': top_5_features}.items():
    X_train_subset, X_test_subset = X_train[features], X_test[features]
    for model_name, model in models.items():
        model.fit(X_train_subset, y_train)
        y_pred = model.predict(X_test_subset)
        accuracy = accuracy_score(y_test, y_pred)
        precision = precision_score(y_test, y_pred, average='weighted', zero_division=1)
        recall = recall_score(y_test, y_pred, average='weighted', zero_division=1)
        results.append([model_name, feature_set, accuracy, precision, recall])

        print(f"Model: {model_name} ({feature_set})")
        print(f"Accuracy: {accuracy}")
        print(f"Precision: {precision}")
        print(f"Recall: {recall}")
        print("Classification Report:")
        print(classification_report(y_test, y_pred))
        print("-" * 50)

        plot_confusion_matrix(y_test, y_pred, model_name, feature_set)
        # Plot precision-recall curve
        if hasattr(model, "predict_proba"):
            y_scores = model.predict_proba(X_test_subset)
            plt.figure(figsize=(10, 6))
            for i in range(len(np.unique(y_test))):
                precision, recall, _ = precision_recall_curve(y_test == i, y_scores[:, i])
                plt.plot(recall, precision, lw=2, label=f'Class {i}')
            plt.xlabel("Recall")
            plt.ylabel("Precision")
            plt.title(f'Precision-Recall Curve: {model_name} ({feature_set})')
            plt.legend()
            plt.show()

# Plot bar charts for accuracy, precision, and recall
# results_df = pd.DataFrame(results, columns=['Model', 'Feature Set', 'Accuracy', 'Precision', 'Recall'])
# fig, axes = plt.subplots(1, 3, figsize=(18, 5))
# sns.barplot(data=results_df, x='Model', y='Accuracy', hue='Feature Set', ax=axes[0])
# axes[0].set_title('Model Accuracy')
# sns.barplot(data=results_df, x='Model', y='Precision', hue='Feature Set', ax=axes[1])
# axes[1].set_title('Model Precision')
# sns.barplot(data=results_df, x='Model', y='Recall', hue='Feature Set', ax=axes[2])
# axes[2].set_title('Model Recall')
# plt.show()
results_df = pd.DataFrame(results, columns=['Model', 'Feature Set', 'Accuracy', 'Precision', 'Recall'])
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
sns.barplot(data=results_df, x='Model', y='Accuracy', hue='Feature Set', ax=axes[0])
axes[0].set_title('Model Accuracy')
axes[0].tick_params(axis='x', rotation=90)
sns.barplot(data=results_df, x='Model', y='Precision', hue='Feature Set', ax=axes[1])
axes[1].set_title('Model Precision')
axes[1].tick_params(axis='x', rotation=90)
sns.barplot(data=results_df, x='Model', y='Recall', hue='Feature Set', ax=axes[2])
axes[2].set_title('Model Recall')
axes[2].tick_params(axis='x', rotation=90)
plt.show()


# Plot feature importance
plt.figure(figsize=(10, 5))
top_features = feature_importance[:10]
top_features.plot(kind='bar')
plt.title('Top 10 Features Importance')
plt.show()

cv_results = {}
for model_name, model in models.items():
    scores = cross_val_score(model, X, y, cv=5, scoring='accuracy')
    cv_results[model_name] = scores
    plt.plot(range(1, 6), scores, marker='o', label=model_name)

plt.xlabel('Fold')
plt.ylabel('Accuracy')
plt.title('Cross-validation Accuracy Curves')
plt.legend()
plt.show()

results_df = pd.DataFrame(results, columns=['Model', 'Feature Set', 'Accuracy', 'Precision', 'Recall'])

best_model_index = results_df[['Accuracy', 'Precision', 'Recall']].mean(axis=1).idxmax()
best_model = results_df.loc[best_model_index]

print("\nBest Model (Averaging Metrics):")
print(best_model)

```
